```python
preds = pd.read_csv("predictions.csv")
preds.drop(["Unnamed: 0", "DateOfBirth"], axis = 1, inplace = True)
preds
```

```python
predsNonPremium = preds.groupby("PremiumPupil").mean().iloc[0]
predsPremium = preds.groupby("PremiumPupil").mean().iloc[1]
premium = preds[preds["PremiumPupil"] == 1]
nonPremium = preds[preds["PremiumPupil"] == 0]
```

```python
print(len(nonPremium))
print(len(premium))
print(len(nonPremium) / len(premium))
print("\n")
print(nonPremium["IsCorrect"].mean())
print(premium["IsCorrect"].mean())
print(nonPremium["IsCorrect"].mean() / premium["IsCorrect"].mean())
```
