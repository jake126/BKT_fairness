```python
preds = pd.read_csv("predictions.csv")
preds.drop(["Unnamed: 0", "DateOfBirth"], axis = 1, inplace = True)
```

```python
times = []
for i in range(len(preds)):
  times.append(date.datetime.strptime(preds.iloc[i]["DateAnswered"], '%Y-%m-%d %H:%M:%S.%f'))
```

```python
preds.drop(["DateAnswered"], axis = 1, inplace = True)
preds["DateAnswered"] = times
```

```python
preds
```

```python
length = len(preds)
timeLimit = date.timedelta(days= 0, hours = 0, minutes = 10)
timeToAnswer = [-1]

for i in range(1, (length)):
  timeNow = preds.iloc[i]["DateAnswered"]
  timePrev = preds.iloc[i - 1]["DateAnswered"]
  timeTaken = timeNow - timePrev
  if (timeTaken > timeLimit):
    timeToAnswer.append(-1)
  else:
    timeToAnswer.append(timeTaken.total_seconds() / 60)

#timeToAnswer
```

```python
preds["TimeToAnswer"] = timeToAnswer
preds
```

```python
timeToAnswers = preds["TimeToAnswer"].tolist()
print(len(timeToAnswers))
```

```python
length = len(timeToAnswers)

def CountHotness(limit):
  runningHot = []
  outer = True
  inner = True
  index = 0

  while (outer == True):
    inner = True
    count = 1
    miniCount = 1

    while (inner == True):
      if ((index + 1) >= length):
        inner = False
        outer = False
      else:
        current = timeToAnswers[index]
        next = timeToAnswers[index + 1]
        index += 1

        if (next == -1):
          inner = False

        elif (next == 0):
          count += 1
          miniCount += 1

        elif ((next == 1) and (miniCount >= limit)):
          count += 1
          miniCount = 1

        else:
          inner = False

    if (count >= limit):
      #runningHot.append(0)
      for i in range(count):
        runningHot.append(1)
    else:
      for i in range(count):
        runningHot.append(0)

  return runningHot

hot2 = CountHotness(2)
hot3 = CountHotness(3)
hot4 = CountHotness(4)
hot5 = CountHotness(5)
```
